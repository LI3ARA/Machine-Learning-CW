```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("spambase.csv")
data.head()
data =data.drop_duplicates() #drop duplicates\

X = data.iloc[:, 0:-1].values  #get all the rows of all the columns except last one to X data set #Features
y = data.iloc[:, -1].values  # get the y data set by taking the lst column of all the rows # Target variables

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, shuffle=True)

from sklearn.preprocessing import StandardScaler

ssc = StandardScaler()
X_train_ssc = ssc.fit_transform(X_train)
X_test_ssc = ssc.transform(X_test)

from sklearn.decomposition import PCA
import numpy as np
# Reduce dimensionality using PCA
pca= PCA()
X_train_pca = pca.fit_transform(X_train_ssc)
X_test_pca = pca.transform(X_test_ssc)
X_ssc = ssc.fit_transform(X)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
knn = KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
```

```python
knn.fit(X_train, y_train)
y_knn_predic_n = knn.predict(X_test)
knn.score(X_test_pca, y_knn_predic_n)
```

```python
knn.fit(X_train_ssc, y_train)
y_knn_predic_scc = knn.predict(X_test_ssc)
knn.score(X_test_pca, y_knn_predic_scc)
```

```python
knn.fit(X_train_pca, y_train)
y_knn_predic_pcc = knn.predict(X_test_pca)
knn.score(X_test_pca, y_knn_predic_pcc)
```

```python
# Define the parameter grid to search over -decision tree
from sklearn.model_selection import GridSearchCV
param_grid = {
    'max_depth': [1, 2, 3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

# Define the grid search object
grid = GridSearchCV(data, param_grid=param_grid, cv=5, scoring='accuracy')

# Fit the grid search object to the data
grid.fit(X_train, y_train)

# Print the best parameters and corresponding score
print(f"Best parameters: {grid.best_params_}")
print(f"Best score: {grid.best_score_}")
```

```python
# grid search

from matplotlib import pyplot
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
model = XGBClassifier()
# encode string class values as integers
label_encoded_y = y
max_depth = range(1, 11, 2)
print(max_depth)
```

```python
n_estimators = range(50, 400, 50)
param_grid = dict(n_estimators=n_estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)
grid_search = GridSearchCV(model, param_grid, scoring="neg_log_loss", n_jobs=-1, cv=kfold)
```

```python
grid_result = grid_search.fit(X, y)
```

```python
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

```python
print(len(max_depth), len(means), len(stds))
```

```python
points = [(depth, mean, std) for depth, mean, std in zip(max_depth, means, stds) if mean is not None and std is not None]
depths, means, stds = zip(*points)

pyplot.errorbar(depths, means, yerr=stds)
pyplot.title("XGBoost max_depth vs Log Loss")
pyplot.xlabel('max_depth')
pyplot.ylabel('Log Loss')
pyplot.savefig('max_depth.png')
```
